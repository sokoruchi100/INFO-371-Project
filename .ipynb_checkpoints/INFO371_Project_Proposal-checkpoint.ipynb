### 1. Introduction
---
*(Introduce the project and describe the objectives. This section will provide an overview of the entire project including the description of the data sets and the specific data analytics methods and techniques the team will use for analyzing the data to address the problem.)*

The project's goal is to create a text classification model that can detect whether a string of text was generated by a human or an AI. The dataset is called the Human vs. LLM Text Corpus, and contains 5 columns listing the text, source, prompt_id, text_length, and word_count. The source will be used as the label for the classification task. The dataset contains 788922 rows of sample text. We will test various text transformation methods such as binary bag of words, term frequency, tfidf, word2vec. We will then combine that with various classifier models like KNN, Decision Tree, Random Forest, Naive Bayes, and a Deep Neural Network. If we have time, we will also attempt to implement a Transformer-based Classification model. We will also use Cross Validation to evaluate the models and use Grid Search to select parameters. The dataset will be split into a training and test set beforehand, with the test set used to evaluate the model with the highest Cross Validation score, while also being used as little as possible to avoid data leakage.

### 2. Problem Definition
---
*(Define the problem that will be solved in this data analytics project. Where does the problem come from? Who would be benefited from the solutions presented in this project? Use formal languages and notations to define the problem as much as possible.)*

The problem is that AI generated text is hard to distinguish between human generated text. As LLMs get more and more advanced, many people, such as students, use AI for writing. Schools and other industries that need to distinguish human vs AI generated works face this challenging issue. Our project seeks to solve this problem by creating a model that can determine text as either human or AI generated.

### 3. Data Sources
---
*(Describe the origin of the data sources. What is the format of the original data? How to access the data?)*

The data set we are using comes from Kaggle: https://www.kaggle.com/datasets/starblasters8/human-vs-llm-text-corpus/data?select=data.csv
It contains over 700k rows of data, each row containing features like text samples, the source of the data, prompt id, text length, and word count. This data is formatted in a csv file and can be easily downloaded from kaggle. The creator of the dataset is Zachary Grinberg, who is writing a research paper on AI Detection and its accuracy/effectiveness. The text samples come from various users as well as LLMs and have a large variety of topics.
