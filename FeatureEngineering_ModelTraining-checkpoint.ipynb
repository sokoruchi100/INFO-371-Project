```python
# import packages
import pandas as pd
import numpy as np

# import the classifiers
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import MultinomialNB

# import count vectorizer and tfidf vectorizer
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

# import the grid search modules
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import f1_score

# import matplotlib
import matplotlib.pyplot as plt
```

```python
# change the path to where you stored the csv
df_train = pd.read_csv("train.csv")
df_test = pd.read_csv("test.csv")
```

```python
stop_words = {"the", "is", "in", "at", "of", "and", "a", "to", "it", "for", "on", "with", "as", "this", "that"}

# adding two additional features
def get_features(text):
    words = str(text).lower().split()
    total_words = len(words)
    if total_words == 0:
        return 0, 0
    # feature 1: vocab richness: how many unique words were used in the text
    richness = len(set(words)) / total_words
    # feature 2: stop word ratio: the number of stop words / total number of words
    stop_ratio = len([w for w in words if w in stop_words]) / total_words
    
    return richness, stop_ratio

# apply the feature function to both dataframes
df_train[['vocab_richness', 'stop_ratio']] = df_train['text'].apply(lambda x: pd.Series(get_features(x)))
df_test[['vocab_richness', 'stop_ratio']] = df_test['text'].apply(lambda x: pd.Series(get_features(x)))

features = ['vocab_richness', 'stop_ratio']
```

```python
# define the parameters and set up the grid search
# combine 4 different models with 3 different vectorizers based on 1) text column only and 2) text column with the two feature columns
models = {
    'KNN': (KNeighborsClassifier(), {'n_neighbors': [3, 5, 11]}),
    'DecisionTree': (DecisionTreeClassifier(), {'max_depth': [None, 10, 20]}),
    'RandomForest': (RandomForestClassifier(), {'n_estimators': [50, 100]}),
    'NaiveBayes': (MultinomialNB(), {'alpha': [0.1, 1.0]})
}

vectorizers = {
    'Binary': CountVectorizer(binary=True),
    'Counts': CountVectorizer(),
    'TF-IDF': TfidfVectorizer()
}

results = []

# loop through all possible combinations
for vector_name, vector_obj in vectorizers.items():
    # vectorize the text column
    X_train_text = v_obj.fit_transform(df_train['text'])
    X_test_text = v_obj.transform(df_test['text'])
    
    # classification based on text column and features
    X_train_combined = np.hstack([X_train_text.toarray(), df_train[features].values])
    X_test_combined = np.hstack([X_test_text.toarray(), df_test[features].values])

    # define the target
    y_train = df_train['ai_generated']
    y_test = df_test['ai_generated']

    for model_name, (model_obj, parameters) in models.items():
        # test 1: text only
        grid_text = GridSearchCV(model_obj, parameters, cv=3, scoring='f1')
        grid_text.fit(X_train_text, y_train)
        f1_text = f1_score(y_test, grid_text.predict(X_test_text))
        
        # test 2: combining text and features
        grid_combined = GridSearchCV(model_obj, parameters, cv=3, scoring='f1')
        grid_combined.fit(X_train_combined, y_train)
        f1_combined = f1_score(y_test, grid_combined.predict(X_test_combined))
        
        results.append({
            'Model': model_name,
            'Vectorizer': vector_name,
            'F1 (Text Only)': round(f1_text, 4),
            'F1 (Text + Features)': round(f1_combined, 4),
            'Best Params': grid_combined.best_params_
        })

# Display Results
final_report = pd.DataFrame(results)
print(final_report.sort_values(by='F1 (Text + Features)', ascending=False))
```

```python
# create a combined label for the model and the vectorizer
final_report['Combo'] = final_report['Model'] + "_" + final_report['Vectorizer']

# sort the report so the best performing models are on the left
plot_df = final_report.sort_values(by='F1 (Text + Features)', ascending=False)

# setup the bar chart positions
x = np.arange(len(plot_df))
width = 0.35

# create the plot
fig = plt.figure(figsize=(14, 7))
ax = fig.add_subplot(1, 1, 1)
rects1 = ax.bar(x - width/2, plot_df['F1 (Text Only)'], width, label='Text Only', color='#3498db')
rects2 = ax.bar(x + width/2, plot_df['F1 (Text + Features)'], width, label='Text + Features', color='#e74c3c')

# add text for labels, title, and x-axis tick labels
ax.set_ylabel('F1 Score')
ax.set_title('Comparison of Prediction Performance')
ax.set_xticks(x)
ax.set_xticklabels(plot_df['Combo'], rotation=45, ha='right')
ax.legend()

# add a grid for easier reading
ax.yaxis.grid(True, linestyle='--', alpha=0.7)

plt.show()
```
